```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
brooklyn_one_bed = pd.read_csv('brooklyn.csv')
brooklyn_one_bed
```

```python
manhattan_one_bed = pd.read_csv('manhattan.csv')
manhattan_one_bed
```

```python
queens_one_bed = pd.read_csv('queens.csv')
queens_one_bed
```

```python
# Selecting the rent column
```

```python
brooklyn_rent_price = brooklyn_one_bed.rent
brooklyn_rent_price
```

```python
manhattan_rent_price = manhattan_one_bed.rent
manhattan_rent_price
```

```python
queens_rent_price = queens_one_bed.rent
queens_rent_price
```

```python
# Finding the mean price of Brooklyn apartments
brooklyn_mean_price = np.average(brooklyn_rent_price)
brooklyn_mean_price
```

```python
# Finding the mean price of Manhattan apartments
manhattan_mean_price = np.average(manhattan_rent_price)
manhattan_mean_price
```

```python
# Finding the mean price of Queens apartments
queens_mean_price = np.average(queens_rent_price)
queens_mean_price
```

```python
# Finding the median price of Brooklyn apartments
brooklyn_median_price = np.median(brooklyn_rent_price)
brooklyn_median_price
```

```python
# Finding the median price of Manhattan apartments
manhattan_median_price = np.median(manhattan_rent_price)
manhattan_median_price
```

```python
# Finding the median price of Queens apartments
queens_median_price = np.median(queens_rent_price)
queens_median_price
```

```python
# Finding the mode price of Brooklyn apartments
brooklyn_mode_price = stats.mode(brooklyn_rent_price)
brooklyn_mode_price
```

```python
# Finding the mode price of Manhattan apartments
manhattan_mode_price = stats.mode(manhattan_rent_price)
manhattan_mode_price
```

```python
# Finding the mode price of Queens apartments
queens_mode_price = stats.mode(queens_rent_price)
queens_mode_price 
```
